```python
# 1. For the suburb Altona, it is postulated that a typical property sells for $800,000.
#Use the data at hand to test this assumption. Is the typical property price really $800,000 or has it increased? 
#Use a significance level of 5%.
```

```python
df_sub_Altona = df[df['Suburb'] == 'Altona']
df_sub_Altona
```

```python
mean_price_altona = df_sub_Altona['Price'].mean()
mean_price_altona
```

```python
#H0 = The typical property price in Altona is $800,000;
#Ha = the price has increased above $800,000
```

```python
null_hypo_value = 800000
t_stat, p_value = stats.ttest_1samp(df_sub_Altona['Price'], null_hypo_value)

print("T-Statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Property price has increased from $800,000")
else:
    print("Property price is not different from $800,000")      
```

```python
# p-value is greater than significance level, therefore Property price has not increased from $800,000.
```

```python
# 2. For the year 2016, is there any difference in prices of properties sold in the summer months vs winter months? 
#Consider months from October till March as winter months and rest as summer months. 
#Use a significance level of 5%. 
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce')
```

```python
df['Date'].info()
```

```python
property_2016 = df[df['Date'].dt.year == 2016]
```

```python
property_2016
```

```python
sum_months = (property_2016['Date'].dt.month >= 4) & (property_2016['Date'].dt.month <= 9)
win_months = ~sum_months
```

```python
summer_properties = property_2016[sum_months]
winter_properties = property_2016[win_months]
```

```python
#Ho = Average property price in summer and winter months are same.
#Ha = Average property prices between summer and winter month are different
```

```python
t_statistic, p_value = stats.ttest_ind(summer_properties['Price'], winter_properties['Price'])
print("T-Statistic:", t_statistic)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference in property prices between summer and winter months.")
else:
    print("There is no significant difference in property prices between summer and winter months.")
```

```python
#As the p-value is lesser than the significant value, we reject the null hypothesis.
```
